7. Generate a dataset of random numbers. Visualize the distribution of a numerical variable

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

np.random.seed(42)

n_samples = 1000
data = np.random.normal(loc=0, scale=1, size=n_samples)

import pandas as pd
df = pd.DataFrame({'Value': data})

plt.figure(figsize=(12, 6))

plt.subplot(1, 2, 1)
plt.hist(df['Value'], bins=30, edgecolor='black', alpha=0.7)
plt.title('Histogram of Random Data')
plt.xlabel('Value')
plt.ylabel('Frequency')

plt.subplot(1, 2, 2)
sns.kdeplot(df['Value'], shade=True)
plt.title('Density Plot of Random Data')
plt.xlabel('Value')
plt.ylabel('Density')

plt.tight_layout()
plt.show()
```

10. Generate a synthetic dataset with correlated features. Visualize the correlation matrix of a dataset using a heatmap

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

np.random.seed(0)
n_samples = 100

feature_A = np.random.normal(loc=50, scale=10, size=n_samples)

feature_B = feature_A + np.random.normal(loc=0, scale=5, size=n_samples)

feature_C = 0.5 * feature_A + np.random.normal(loc=0, scale=7, size=n_samples)

df = pd.DataFrame({
    'Feature_A': feature_A,
    'Feature_B': feature_B,
    'Feature_C': feature_C
})

correlation_matrix = df.corr()

plt.figure(figsize=(8, 6))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt='.2f', vmin=-1, vmax=1)

plt.title('Correlation Matrix Heatmap')
plt.show()
```

13. Using the sales data, generate a heatmap to visualize the variation in sales across different months and
days.

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Set seed for reproducibility
np.random.seed(20)

# Create the data
data = {
    'Month': np.random.choice(['Jan', 'Feb', 'Mar', 'Apr', 'May'], 100),
    'Day': np.random.choice(range(1, 31), 100),
    'Sales': np.random.randint(1000, 5000, 100)
}
df = pd.DataFrame(data)

# Aggregate the sales data by Month and Day
sales_aggregated = df.groupby(['Month', 'Day'])['Sales'].sum().reset_index())

# Pivot the DataFrame to create a matrix for the heatmap
heatmap_data = sales_aggregated.pivot ('Day', 'Month', 'Sales')

# Create the heatmap
plt.figure(figsize=(12, 8))
sns.heatmap(heatmap_data, cmap='YlGnBu', annot=True, fmt='g', linewidths=.5)
plt.title('Sales Heatmap by Month and Day')
plt.xlabel('Month')
plt.ylabel('Day')
plt.xticks(rotation=45)
plt.show()
```

14. Using the given x and y data, generate a 3D surface plot to visualize the function z = sin(x + y)

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

x = np.linspace(-5, 5, 100)
y = np.linspace(-5, 5, 100)
x, y = np.meshgrid(x, y)
z = np.sin(np.sqrt(x**2 + y**2))

data = {
    'X': x.flatten(),
    'Y': y.flatten(),
    'Z': z.flatten()
}
df = pd.DataFrame(data)

fig = plt.figure(figsize=(10, 8))
ax = fig.add_subplot(111, projection='3d')

ax.plot_surface(x, y, z, cmap='viridis', edgecolor='none')

ax.set_xlabel('X axis')
ax.set_ylabel('Y axis')
ax.set_zlabel('Z axis')
ax.set_title('3D Surface Plot of z = sin(sqrt(x + y))')

plt.show()
```

15.  Using the given dataset, create a bubble chart to represent each country's population (y-axis), GDP (xaxis), and bubble size proportional to the population

16. Create a Bokeh plot displaying a sine wave. Set x-values from 0 to 10 and y-values as the sine of x.

```python
from bokeh.plotting import figure, show
from bokeh.io import output_notebook
import numpy as np

output_notebook()

x = np.linspace(0, 10, 100)
y = np.sin(x)


p = figure(title="Sine Wave", x_axis_label='x', y_axis_label='sin(x)', width=800, height=400)

p.line(x, y, legend_label="sin(x)", line_width=2, color='blue')

show(p)
```

17. Create a Bokeh scatter plot using randomly generated x and y values. Use different sizes and colors for the
markers based on the 'sizes' and 'colors' columns

```python
import numpy as np
import pandas as pd
from bokeh.plotting import figure, show
from bokeh.io import output_notebook

output_notebook()

np.random.seed(42)
n = 100
x = np.random.rand(n) * 100
y = np.random.rand(n) * 100
sizes = np.random.randint(10, 50, n)
colors = np.random.choice(['red', 'green', 'blue', 'orange', 'purple'], n)

data = pd.DataFrame({'x': x, 'y': y, 'sizes': sizes, 'colors': colors})

p = figure(title="Random Scatter Plot", x_axis_label='X', y_axis_label='Y', width=800, height=400)

p.scatter('x', 'y', size='sizes', color='colors', source=data, legend_field='colors', alpha=0.6)

show(p)
```

18.  Generate a Bokeh bar chart representing the counts of different fruits using the following dataset

```python
from bokeh.plotting import figure, show
from bokeh.io import output_notebook
from bokeh.transform import factor_cmap

output_notebook()

fruits = ['Apples', 'Oranges', 'Bananas', 'Pears']
counts = [20, 25, 30, 35]

p = figure(x_range=fruits, title="Fruit Counts",
           x_axis_label='Fruits', y_axis_label='Counts',
           width=600, height=400)

p.vbar(x=fruits, top=counts, width=0.5, color=factor_cmap('x', palette=['#718dbf', '#e84d60', '#f5bb00', '#3cba0d'], factors=fruits))

for fruit, count in zip(fruits, counts):
    p.text(x=fruit, y=count + 1, text=[str(count)], text_align='center')

show(p)
```

19.  Create a Bokeh histogram to visualize the distribution of the given data

```python
import numpy as np
from bokeh.plotting import figure, show
from bokeh.io import output_notebook

output_notebook()

data_hist = np.random.randn(1000)

hist, edges = np.histogram(data_hist, bins=30)

p = figure(title="Histogram of Random Data",
           x_axis_label='Value',
           y_axis_label='Frequency',
           width=800, height=400)

p.quad(top=hist, bottom=0, left=edges[:-1], right=edges[1:],
       fill_color='blue', line_color='black', alpha=0.7)

show(p)
```

20. Create a Bokeh heatmap using the provided dataset

```python
import numpy as np
from bokeh.plotting import figure, show
from bokeh.io import output_notebook
from bokeh.transform import linear_cmap
from bokeh.models import ColorBar, LinearColorMapper

output_notebook()

data_heatmap = np.random.rand(10, 10)
x = np.linspace(0, 1, 10)
y = np.linspace(0, 1, 10)
xx, yy = np.meshgrid(x, y)

heatmap_data = []
for i in range(data_heatmap.shape[0]):
    for j in range(data_heatmap.shape[1]):
        heatmap_data.append((xx[i, j], yy[i, j], data_heatmap[i, j]))

p = figure(title="Heatmap", x_axis_label='X', y_axis_label='Y', width=600, height=600)

mapper = LinearColorMapper(palette="Viridis256", low=data_heatmap.min(), high=data_heatmap.max())

p.rect(x='x', y='y', width=0.1, height=0.1, source=heatmap_data,
       line_color=None, fill_color=linear_cmap('value', palette="Viridis256", low=data_heatmap.min(), high=data_heatmap.max()))

color_bar = ColorBar(color_mapper=mapper, location=(0, 0))
p.add_layout(color_bar, 'right')

show(p)
```
